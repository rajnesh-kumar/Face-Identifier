```python
import numpy as np
import cv2
import pickle

face_cascade = cv2.CascadeClassifier("OpenCV-Python-Series-master\src\cascades\data\haarcascade_frontalface_alt2.xml")
#This is a face classifier... To use face classifier we must use grey scale

eye_cascade = cv2.CascadeClassifier("OpenCV-Python-Series-master\src\cascades\data\haarcascade_eye.xml")

smile_cacade = cv2.CascadeClassifier("OpenCV-Python-Series-master\src\cascades\data\haarcascade_smile.xml")



labels = {}

recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("trainner.yml")
with open("labels.pickle", "rb") as f:     #rb means reading bytes
    og_labels = pickle.load(f)
    labels = {v:k for k,v in og_labels.items()} #inverting the dictionary. key, value becomes value, key
    
cap = cv2.VideoCapture(0)   #It won't capture the video it will just keep the camera on.

while(True):     #we use the loop because we want the camera to be continously on until we break the loop.
    # Capture frame-by-frame
    ret, frame = cap.read()             #This is reading frame by frame

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) #cvtColor will convert the color, here we're converting to grey
    faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.5, minNeighbors = 5) #ScaleFacetpr is for accuracy
    for (x, y, w, h) in faces: #iterating through the faces
        #print(x, y, w, h)
        roi_gray = gray[y:y+h, x:x+w] #roi means region of interest which is just the face in the whole frame
        roi_color = frame[y:y+h, x:x+w] #(ycord start, ycord+height) #(xcord start, xcord+width)
        
        
        #Recognizer
        
        
        id_, conf = recognizer.predict(roi_gray)
        if conf >=45: #and conf <=85:
            print(id_)
            print(labels[id_])
            font = cv2.FONT_HERSHEY_SIMPLEX
            name = labels[id_]
            color = (255, 255, 255)
            stroke = 2
            cv2.putText(frame, name, (x,y), font, 1, color, stroke, cv2.LINE_AA)
        #Now load label names from pickle
        
        img_item = "my-image.png"
        cv2.imwrite(img_item, roi_gray)     #this will save the image, not the whole frame but only the face part
        
        
        #Display the rectangle
        color = (255, 0, 0) #BGR 0-255 This is going to be totally blue because the other two colors are set to 0
        stroke = 2 #How thick do you want the line to be
        end_cord_x = x + w #These are not actually the height and width, these are the coordinates
        end_cord_y = y + h
        cv2.rectangle(frame, (x,y), (end_cord_x, end_cord_y), color, stroke) #the command is rectangle but it can turn out
        #to be a square also, or anything else too.
        
        eyes = eye_cascade.detectMultiScale(roi_gray)
        
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey + eh), (0, 255, 0), 2)
        
    # Display the resulting frame
    cv2.imshow('frame',frame)
    #cv2.imshow('gray',gray)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
